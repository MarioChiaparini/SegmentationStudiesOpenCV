## i) Importing libraries and Images

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
import cv2
```

```python
amostra_imagem = cv2.imread('./images/cube.jpg')
```

```python
imagem = cv2.cvtColor(amostra_imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem)
```

## ii) Apply Otsu Thresholding on Image

```python
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_RGB2GRAY)
thresh = threshold_otsu(imagem_gray)
imagem_otsu = imagem_gray < thresh
```

```python
plt.imshow(imagem_otsu)
```

## iii) Segmentation Process

```python
def filter_image(image,mask):
    r = image[:,:,0] * mask 
    g = image[:,:,1] * mask
    b = image[:,:,2] * mask
    return np.dstack([r,g,b])
```

```python
filtered = filter_image(imagem, imagem_otsu)
```

```python
plt.imshow(filtered)
```
