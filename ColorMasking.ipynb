## i) Importing libraries and Images

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
import cv2
```

```python
amostra_imagens = cv2.imread('./images/pokemon.jpg')
imagem = cv2.cvtColor(amostra_imagens, cv2.COLOR_BGR2RGB)
```

```python
plt.axis('off')
plt.imshow(imagem)
```

## ii) Create Mask by Detecting Color

```python
low = np.array([0,0,0])
high = np.array([215,51,51])
```

```python
mask = cv2.inRange(imagem, low, high)
```

```python
plt.axis('off')
plt.imshow(mask)
```

## iii) Apply the Mask

```python
resultado = cv2.bitwise_and(imagem, imagem, mask=mask)
```

```python
plt.imshow(resultado)
```
